# Main Difference

## Random Forest Regressor

Each Decision Tree predicts a number.

Example

```
Tree 1 â†’ 52

Tree 2 â†’ 55

Tree 3 â†’ 50

Tree 4 â†’ 54
```

Final Prediction

```
Average

â†“

52.75
```

---

## Random Forest Classifier

Each Decision Tree predicts a class.

Example

```
Tree 1 â†’ Spam

Tree 2 â†’ Spam

Tree 3 â†’ Not Spam

Tree 4 â†’ Spam

Tree 5 â†’ Spam
```

Final Prediction

```
Majority Voting

â†“

Spam
```

# Workflow

```
Training Data

â†“

Bootstrap Sampling

â†“

Build Multiple Decision Trees

â†“

Each Tree Predicts a Class

â†“

Majority Voting

â†“

Final Class Prediction
```

# Interview Questions

### 1. What is the main difference between Random Forest Regressor and Random Forest Classifier?

The regressor averages the predictions from all trees, while the classifier uses **majority voting** to predict the final class.

---

### 2. Which splitting criteria are used in Random Forest Classification?

- Gini Impurity
- Entropy
- Log Loss

---

### 3. Does Random Forest Classifier support probability prediction?

Yes.

Using:

```python
model.predict_proba(X_test)
```

---

### 4. Does Random Forest Classifier require feature scaling?

No.

Like Decision Trees, it is based on recursive splitting rather than distance calculations.

---

### 5. What is the default splitting criterion?

```python
criterion="gini"
```
