# Gradient Boosting Classifier (Differences from Gradient Boosting Regressor)

> **Prerequisite:** Read the Gradient Boosting Regressor notes first. The concepts of boosting, sequential learning, weak learners, residual learning, learning rate, and ensemble construction are the same.

---

# What Changes?

Gradient Boosting Regressor predicts a **continuous value**.

Example

```
House Price

↓

₹48,50,000
```

Gradient Boosting Classifier predicts a **class label**.

Example

```
Customer

↓

Will Churn

Yes
```

---

# Main Difference

## Gradient Boosting Regressor

Each tree learns the **residual (error)** between the actual and predicted values.

Loss functions:

- Squared Error
- Absolute Error
- Huber
- Quantile

Final output:

```
Continuous Value
```

---

## Gradient Boosting Classifier

Each tree learns to reduce the **classification error** by minimizing a **classification loss function**.

Final output:

```
Probability

↓

Class Label
```

---

# Classification Workflow

```
Training Data

↓

Initial Probability

↓

Calculate Classification Loss

↓

Train Small Decision Tree

↓

Update Predictions

↓

Repeat Sequentially

↓

Final Probability

↓

Class Prediction
```

---

# Loss Function

The most common loss function is **Log Loss (Deviance)**.

Binary Classification:

$$
Loss = -\left[y\log(p)+(1-y)\log(1-p)\right]
$$

Where:

- \(y\) = Actual class (0 or 1)
- \(p\) = Predicted probability of class 1

The objective is to **minimize Log Loss**.

# Interview Questions

### 1. What is the main difference between Gradient Boosting Regressor and Classifier?

The regressor minimizes regression loss (e.g., MSE), while the classifier minimizes **classification loss (Log Loss)**.

---

### 2. What is the default loss function?

```python
loss="log_loss"
```

---

### 3. Does Gradient Boosting Classifier support probability prediction?

Yes.

```python
model.predict_proba(X_test)
```

---

### 4. Does Gradient Boosting Classifier require feature scaling?

No.

Like Decision Trees, it is tree-based and does not rely on distance calculations.

---

### 5. What is the role of `learning_rate`?

It controls how much each new tree contributes to correcting the errors made by previous trees.
