# XGBoost Classifier (Differences from XGBoost Regressor)

> **Prerequisite:** Read the XGBoost Regressor notes first. Concepts such as Gradient Boosting, Second-Order Optimization, Regularization (L1/L2), Pruning, Missing Value Handling, Parallel Processing, and Feature Importance are the same.

---

# What Changes?

XGBoost Regressor predicts a **continuous value**.

Example

```
House Price

↓

₹65,00,000
```

XGBoost Classifier predicts a **class label**.

Example

```
Loan

↓

Approved
```

---

# Main Difference

## XGBoost Regressor

Predicts

```
Continuous Value
```

Objective

```python
objective="reg:squarederror"
```

Loss Function

```
Mean Squared Error
```

---

## XGBoost Classifier

Predicts

```
Class Label
```

Objective

```python
objective="binary:logistic"
```

Loss Function

```
Binary Log Loss
```

Output

```
Probability

↓

Class Label
```

---

# Classification Workflow

```
Training Data

↓

Initial Prediction

↓

Calculate Log Loss

↓

Train New Tree

↓

Correct Previous Errors

↓

Repeat

↓

Final Probability

↓

Class Prediction
```

---

# Classification Objectives

## Binary Classification

Used when there are only two classes.

Example

```
Spam

Not Spam
```

Objective

```python
objective="binary:logistic"
```

Output

```
Probability

↓

0 or 1
```

---

## Multi-Class Classification

Used for more than two classes.

Example

```
Cat

Dog

Horse
```

Objective

```python
objective="multi:softprob"
```

Required Parameter

```python
num_class=3
```

Output

```
Probability for each class
```

Alternative

```python
objective="multi:softmax"
```

Returns the predicted class label directly instead of probabilities.

# Interview Questions

### 1. What is the main difference between XGBoost Regressor and XGBoost Classifier?

The regressor predicts continuous values using a regression objective, while the classifier predicts class labels using classification objectives such as `binary:logistic`.

---

### 2. What is the default objective for binary classification?

```python
objective="binary:logistic"
```

---

### 3. What is the difference between `multi:softmax` and `multi:softprob`?

- `multi:softmax` returns the predicted class label.
- `multi:softprob` returns the probability for every class.

---

### 4. Does XGBoost Classifier support probability prediction?

Yes.

```python
model.predict_proba(X_test)
```

---

### 5. Does XGBoost require feature scaling?

No.

Like other tree-based algorithms, XGBoost is not dependent on feature scaling.

---

# Summary

|Feature|Regressor|Classifier|
|--------|----------|-----------|
|Prediction|Continuous Value|Class Label|
|Objective|`reg:squarederror`|`binary:logistic`, `multi:softprob`|
|Loss Function|Regression Loss|Binary/Multiclass Log Loss|
|Output|Numeric Value|Probability → Class|
|Evaluation|MAE, RMSE, R²|Accuracy, Precision, Recall, F1, ROC-AUC|
|Probability Prediction|❌|✅ `predict_proba()`|
