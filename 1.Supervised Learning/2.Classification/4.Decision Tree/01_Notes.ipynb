# Decision Tree Classifier (Differences from Decision Tree Regressor)

> **Prerequisite:** Review the Decision Tree Regression notes first. The concepts of tree construction, recursive splitting, tree depth, pruning, and most parameters are the same.

---

# What Changes?

Decision Tree Regression predicts a **continuous value**.

Example:

```
House Price

↓

₹45,00,000
```

Decision Tree Classifier predicts a **class label**.

Example:

```
Loan

↓

Approved
```

or

```
Email

↓

Spam
```

---

# Main Difference

## Decision Tree Regressor

Goal:

Predict

```
Continuous Values
```

Uses splitting criteria like:

- Mean Squared Error (MSE)
- Friedman MSE
- Mean Absolute Error (MAE)

Leaf Node:

```
Average Target Value
```

---

## Decision Tree Classifier

Goal:

Predict

```
Discrete Classes
```

Uses impurity measures:

- Gini Impurity (default)
- Entropy (Information Gain)
- Log Loss (supported in newer versions of Scikit-Learn)

Leaf Node:

```
Majority Class
```

---

# Classification Workflow

```
Training Data

↓

Find Best Split

↓

Measure Node Impurity

↓

Split Dataset

↓

Repeat Recursively

↓

Leaf Node

↓

Majority Class Prediction
```

---

# Gini Impurity

Default splitting criterion.

Formula:

$$
Gini = 1 - \sum_{i=1}^{n} p_i^2
$$

Where:

- \(p_i\) = Probability of class \(i\)

---

## Example

Suppose a node contains:

```
80 Positive

20 Negative
```

Probabilities:

$$
P(Positive)=0.8
$$

$$
P(Negative)=0.2
$$

Calculation:

$$
Gini = 1-(0.8^2+0.2^2)
$$

$$
=1-(0.64+0.04)
$$

$$
=0.32
$$

Lower Gini means a purer node.

---

# Entropy

Entropy measures the amount of uncertainty in a node.

Formula:

$$
Entropy = -\sum_{i=1}^{n} p_i \log_2(p_i)
$$

Properties:

- Entropy = 0 → Pure node
- Higher entropy → More mixed classes

---

# Information Gain

Information Gain measures how much uncertainty is reduced after a split.

Formula:

$$
Information\ Gain =
Entropy(Parent)
-
Weighted\ Entropy(Children)
$$

The split with the **highest Information Gain** is selected.

---

# Gini vs Entropy

|Feature|Gini|Entropy|
|---|---|---|
|Default in Scikit-Learn|✅|❌|
|Uses Logarithm|❌|✅|
|Training Speed|Faster|Slightly Slower|
|Interpretability|Simple|More Information-Theoretic|

In practice, both often produce similar trees.

# Interview Questions

### 1. What is the main difference between Decision Tree Regressor and Decision Tree Classifier?

Regression predicts continuous values, whereas the classifier predicts discrete class labels.

---

### 2. Which splitting criteria are used in Decision Tree Classification?

- Gini Impurity
- Entropy (Information Gain)
- Log Loss

---

### 3. Which criterion is the default in Scikit-Learn?

`criterion="gini"`

---

### 4. Which is faster: Gini or Entropy?

Gini is generally faster because it does not use logarithmic calculations.

---

### 5. What does a leaf node contain in Decision Tree Classification?

The majority class (or class probabilities), instead of the average target value used in regression.
