# LightGBM Classifier (Differences from LightGBM Regressor)

> **Prerequisite:** Read the LightGBM Regressor notes first. Concepts such as Histogram-based Learning, Leaf-wise Tree Growth, GOSS (Gradient-based One-Side Sampling), EFB (Exclusive Feature Bundling), Boosting Process, and most hyperparameters are identical.

---

# What Changes?

LightGBM Regressor predicts a **continuous value**.

Example

```
House Price

↓

₹72,00,000
```

LightGBM Classifier predicts a **class label**.

Example

```
Customer

↓

Will Buy

Yes
```

---

# Main Difference

## LightGBM Regressor

Predicts

```
Continuous Value
```

Objective

```python
objective="regression"
```

Loss Function

```
Regression Loss
```

---

## LightGBM Classifier

Predicts

```
Class Label
```

Objective

```python
objective="binary"
```

Loss Function

```
Binary Log Loss
```

Output

```
Probability

↓

Class Label
```

---

# Classification Workflow

```
Training Data

↓

Initialize Prediction

↓

Calculate Log Loss

↓

Train New Tree

↓

Update Model

↓

Repeat

↓

Final Probability

↓

Predicted Class
```

# Interview Questions

### 1. What is the main difference between LightGBM Regressor and Classifier?

The regressor predicts continuous values, while the classifier predicts class labels using binary or multiclass objectives.

---

### 2. What is the objective for binary classification?

```python
objective="binary"
```

---

### 3. Which parameter is required for multiclass classification?

```python
num_class
```

It specifies the total number of classes.

---

### 4. Does LightGBM Classifier support probability prediction?

Yes.

```python
model.predict_proba(X_test)
```

---

### 5. Does LightGBM require feature scaling?

No.

It is a tree-based algorithm and is not affected by feature scaling.

# Summary

|Feature|Regressor|Classifier|
|--------|----------|-----------|
|Prediction|Continuous Value|Class Label|
|Objective|`regression`|`binary`, `multiclass`|
|Loss Function|Regression Loss|Binary/Multiclass Log Loss|
|Output|Numeric Value|Probability → Class|
|Evaluation|MAE, RMSE, R²|Accuracy, Precision, Recall, F1, ROC-AUC|
|Probability Prediction|❌|✅ `predict_proba()`|
