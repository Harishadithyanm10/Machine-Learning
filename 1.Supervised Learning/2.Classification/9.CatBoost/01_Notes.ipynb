> **Prerequisite:** Read the CatBoost Regressor notes first. Concepts such as Ordered Boosting, Ordered Target Encoding, Handling Categorical Features, Symmetric Trees, Missing Value Handling, and most hyperparameters are identical.

# What Changes?

CatBoost Regressor predicts a **continuous value**.

Example

```
House Price

↓

₹58,00,000
```

CatBoost Classifier predicts a **class label**.

Example

```
Customer

↓

Will Purchase

Yes
```

---

# Main Difference

## CatBoost Regressor

Predicts

```
Continuous Value
```

Objective

```python
loss_function="RMSE"
```

Output

```
Numeric Value
```

---

## CatBoost Classifier

Predicts

```
Class Label
```

Objective

```python
loss_function="Logloss"
```

Output

```
Probability

↓

Class Label
```

---

# Classification Workflow

```
Training Data

↓

Ordered Boosting

↓

Build Symmetric Trees

↓

Minimize Log Loss

↓

Repeat

↓

Final Probability

↓

Predicted Class
```

# Handling Categorical Features

Just like the regressor, CatBoost Classifier can directly use categorical columns.

Example

```python
model.fit(

    X_train,

    y_train,

    cat_features=[0, 2, 5]

)
```

# Interview Questions

### 1. What is the main difference between CatBoost Regressor and Classifier?

The regressor predicts continuous values using regression loss functions (e.g., RMSE), while the classifier predicts class labels using classification loss functions such as Logloss.

---

### 2. Which loss function is commonly used for binary classification?

```python
loss_function="Logloss"
```

---

### 3. Which loss function is used for multiclass classification?

```python
loss_function="MultiClass"
```

---

### 4. Does CatBoost Classifier support probability prediction?

Yes.

```python
model.predict_proba(X_test)
```

---

### 5. Does CatBoost require feature scaling?

No.

Like other tree-based algorithms, CatBoost is not affected by feature scaling.

---

### 6. Does CatBoost require One-Hot Encoding for categorical features?

No.

It can process categorical features directly using the `cat_features` parameter.
