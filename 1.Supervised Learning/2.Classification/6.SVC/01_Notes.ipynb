# Support Vector Machine (SVC) (Differences from Support Vector Regression)

> **Prerequisite:** Read the SVR (Support Vector Regression) notes first. Concepts such as Hyperplane, Support Vectors, Maximum Margin, Kernel Trick, Linear Kernel, Polynomial Kernel, RBF Kernel, Sigmoid Kernel, and most hyperparameters are identical.

---

# What Changes?

SVR predicts a **continuous value**.

Example

```
House Price

↓

₹45,00,000
```

SVC predicts a **class label**.

Example

```
Tumor

↓

Benign
```

---

# Main Difference

## Support Vector Regression (SVR)

Predicts

```
Continuous Value
```

Uses

```
ε-insensitive loss
```

Output

```
Numeric Value
```

---

## Support Vector Classification (SVC)

Predicts

```
Class Label
```

Uses

```
Maximum Margin Classification
```

Output

```
Probability

↓

Class Label
```

---

# Classification Workflow

```
Training Data

↓

Find Optimal Hyperplane

↓

Maximize Margin

↓

Identify Support Vectors

↓

Train Model

↓

Predict Class
```

---

# Decision Rule

Unlike SVR,

which predicts

```
y = 52.4
```

SVC determines

```
Which side

of the Hyperplane

does the sample belong to?
```

Example

```
Class A

|

Hyperplane

|

Class B
```

If the sample is on the left

↓

Class A

If on the right

↓

Class B

---

# Multi-Class Classification

SVC supports multiclass classification automatically.

Scikit-Learn uses:

```
One-vs-One (OvO)
```

by default.

For 4 classes

```
A

B

C

D
```

Classifiers created:

```
A vs B

A vs C

A vs D

B vs C

B vs D

C vs D
```

The class with the most votes is selected.

# Interview Questions

### 1. What is the main difference between SVR and SVC?

SVR predicts continuous values using an ε-insensitive loss, while SVC predicts class labels by finding the maximum-margin separating hyperplane.

---

### 2. Which parameter enables probability prediction?

```python
probability=True
```

---

### 3. Does SVC require feature scaling?

Yes.

Like SVR, SVC is distance-based and performs best when features are scaled.

---

### 4. Which kernels are commonly used in SVC?

- Linear
- Polynomial
- RBF (Gaussian)
- Sigmoid

---

### 5. Which multiclass strategy does Scikit-Learn use internally?

Training is performed using the **One-vs-One (OvO)** strategy. The `decision_function_shape` parameter controls how decision scores are presented, with `"ovr"` being the default output format.
