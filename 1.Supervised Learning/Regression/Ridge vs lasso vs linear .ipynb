# Linear Regression vs Ridge Regression vs Lasso Regression

| Feature | Linear Regression | Ridge Regression | Lasso Regression |
|---------|-------------------|------------------|------------------|
| Full Name | Ordinary Least Squares (OLS) | Linear Regression + L2 Regularization | Linear Regression + L1 Regularization |
| Purpose | Fit the best linear relationship | Reduce overfitting | Reduce overfitting + Feature Selection |
| Cost Function | $RSS$(Residual Sum of Squares) | $RSS + \alpha \sum \beta^2$ | $RSS + \alpha \sum |\beta|$ |
| Regularization | ❌ None | ✅ L2 | ✅ L1 |
| Penalty | None | Square of coefficients | Absolute value of coefficients |
| Shrinks Coefficients | ❌ No | ✅ Yes | ✅ Yes |
| Makes Coefficients Zero | ❌ No | ❌ No | ✅ Yes |
| Feature Selection | ❌ No | ❌ No | ✅ Yes |
| Handles Multicollinearity | ❌ Poorly | ✅ Excellent | ✅ Good |
| Controls Overfitting | ❌ No | ✅ Yes | ✅ Yes |
| Bias | Low | Medium | Higher |
| Variance | High | Lower | Lower |
| Model Complexity | High | Medium | Low |
| Keeps All Features | ✅ Yes | ✅ Yes | ❌ No |
| Removes Features | ❌ No | ❌ No | ✅ Yes |
| Hyperparameter | None | `alpha` | `alpha` |
| Cross Validation Version | None | `RidgeCV` | `LassoCV` |
| Typical Solver | OLS | Multiple Solvers | Coordinate Descent |
| Scikit-learn Class | `LinearRegression` | `Ridge` | `Lasso` |

# When to Use Which Model?

| Situation | Recommended Model | Reason |
|-----------|-------------------|--------|
| Simple dataset with little noise | **Linear Regression** | No regularization needed |
| Overfitting due to correlated features | **Ridge Regression** | L2 regularization stabilizes coefficients |
| Many irrelevant features | **Lasso Regression** | Automatically removes unimportant features |
| Need automatic tuning of `alpha` | **RidgeCV / LassoCV** | Uses cross-validation to select the best `alpha` |
